```python
import numpy
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn import svm
```

```python
X=[1,5,1.5,8,1,9]
Y=[2,8,1,8,0.6,11]
plt.scatter(X,Y)
plt.show()
```

```python
data=numpy.array(list(zip(X,Y)))
```

```python
data
```

```python
target=[0,1,0,1,0,1]
```

```python
classifier=svm.SVC(kernel="linear",C=1.0)
classifier.fit(data,target)
```

```python
classifier.predict(numpy.array([2.57,2.67]).reshape(1,2))
```

```python
w=classifier.coef_[0]
```

```python
#modelo w0.x+w1.y+e=0
#Ecuacion del hiperplano en 2D y=a.x+b
```

```python
a=-w[0]/w[1]
b=-classifier.intercept_[0]/w[1]
```

```python
xx=numpy.linspace(0,10)
yy=b+a*xx
```

```python
plt.plot(xx,yy,"k-",label="Hiperplano de separacion")
plt.scatter(X,Y,c=target)
plt.legend()
plt.plot()
```
