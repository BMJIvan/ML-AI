```python
import numpy 
import matplotlib.pyplot as plt
```

```python
x=numpy.sort(5*numpy.random.rand(200,1),axis=0)
y=numpy.sin(x).ravel()
y[::5]+=3*(0.5-numpy.random.rand(40))
```

```python
plt.scatter(x,y,color="darkorange",label="data")
```

```python
from sklearn.svm import SVR
```

```python
C=1E3
svr_lin=SVR(kernel="linear",C=C)
svr_cir=SVR(kernel="rbf",C=C,gamma=0.1)
svr_pol=SVR(kernel="poly",C=C,degree=3)
```

```python
y_lin=svr_lin.fit(x,y).predict(x)
y_cir=svr_cir.fit(x,y).predict(x)
y_pol=svr_pol.fit(x,y).predict(x)
```

```python
lw=2
plt.figure(figsize=(16,9))
plt.scatter(x,y,color="darkorange",label="data")
plt.plot(x,y_lin,color="navy",lw=lw,label="kernel svm lineal")
plt.plot(x,y_cir,color="c",lw=lw,label="svm radial")
plt.plot(x,y_pol,color="cornflowerblue",label="svm polinomico")
plt.xlabel("x")
plt.ylabel("y")
plt.title("suppor vector regression")
plt.legend()
plt.show()
```
