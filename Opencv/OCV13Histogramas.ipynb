```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = 1000
img = cv.imread("LEGO_ROSTROS\P1.jpg")
cv.imshow("window", img)
cv.waitKey(t)

gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
cv.imshow("window", gray)
cv.waitKey(t)

blank = np.zeros(img.shape[:2], dtype = "uint8")

x = img.shape[0]//3
y = img.shape[1]//3
circle = cv.circle(blank, (x+50, y+50), 150, 255, -1,)
cv.imshow("window", circle)
cv.waitKey(t)

mask = cv.bitwise_and(gray, gray, mask=circle)
cv.imshow("window", mask)
cv.waitKey(t)

gray_hist = cv.calcHist([mask], [0], None, [256], [0, 256])

cv.destroyAllWindows()
```

```python

plt.figure()
plt.title("histogram")
plt.xlabel("Bins")
plt.ylabel("# of pixels")
plt.plot(gray_hist)
plt.show()
```

```python
cv.imshow("window", img)
cv.waitKey(t)

mask_color = cv.bitwise_and(img, img, mask=circle)
cv.imshow("window", mask_color)
cv.waitKey(t)

color = ["b", "g", "r"]
for i, col in enumerate(color):
    hist = cv.calcHist([mask_color], [i], None, [256], [1, 256])
    plt.plot(hist, color = col)
    plt.xlim([0, 256])

cv.destroyAllWindows()
plt.show()
```
