```python
import cv2 as cv
import sys
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
source = cv.VideoCapture(0)
win_name = 'camera'
cv.namedWindow(win_name, cv.WINDOW_NORMAL)

while cv.waitKey(1) != 27:
    has_frame, frame = source.read()
    if not has_frame:
        break
    cv.imshow(win_name, frame)

source.release()
cv.destroyAllWindows()
```

```python
filename = "./VIDEO/This_is_it.mp4"

video = cv.VideoCapture(filename)
if video.isOpened() == False:
    print("error")
```

```python
capture, frame = video.read()
```

```python
plt.imshow(frame[...,::-1])
```

```python
fps = int(video.get(cv.CAP_PROP_FRAME_COUNT))
frame_width = int(video.get(cv.CAP_PROP_FRAME_WIDTH))
frame_height = int(video.get(cv.CAP_PROP_FRAME_HEIGHT))

#define the codec and create videowriter objetc
out_avi = cv.VideoWriter("out1.avi", cv.VideoWriter_fourcc("M", "J", "P", "G"), fps, (frame_width, frame_height), True)
out_mp4 = cv.VideoWriter("out1.mp4", cv.VideoWriter_fourcc(*'XVID'), fps, (frame_width, frame_height))
```

```python
i = 0
while(video.isOpened()):
    capture, frame = video.read()
    
    if capture == True:
        out_avi.write(frame)
        out_mp4.write(frame)
    else:
        break
    i += 1
    if i == 200:
        break
```

```python
video.release()
out_mp4.release()
out_avi.release()
```
