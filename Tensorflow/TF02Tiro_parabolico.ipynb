```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
dt = 0.01
pos = []
vel = []
pos_init = 1000
vel_init = 100
tiempo = 20
muestras = int(tiempo/dt)
```

```python
A = tf.constant([[0, 1],
                 [0, 0]], dtype=tf.float32)
B = tf.constant([[0],
                 [1]], dtype=tf.float32)
u = tf.constant(-9.81, dtype=tf.float32)
X = tf.Variable([[pos_init],
                 [vel_init]], dtype=tf.float32)
X_p = tf.Variable([[0],
                   [0]], dtype=tf.float32)
```

```python
for _ in range(muestras):
    X_p = tf.add(tf.matmul(A, X), tf.multiply(B, u))
    X = tf.add(X, tf.multiply(X_p, dt))
    pos.append(X.numpy()[0][0])
    vel.append(X.numpy()[1][0])
```

```python
plt.plot(pos)
```

```python
plt.plot(vel)
```
